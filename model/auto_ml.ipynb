```python
import pycaret
```

```python
import pandas as pd
ecfp_final_data = pd.read_csv("Chembl_pIC50_ecfp.csv")
ecfp_final_data.head()
```

```python
ecfp_final_data = ecfp_final_data.drop(["SMILES"],axis=1)
ecfp_final_data.head()
```

```python
ecfp_final_data_x=ecfp_final_data.drop(["pIC50"],axis=1)
ecfp_final_data_y=ecfp_final_data['pIC50']
```

```python
from sklearn.model_selection import train_test_split
ecfp_X_train, ecfp_X_test, ecfp_y_train, ecfp_y_test = train_test_split(ecfp_final_data_x, ecfp_final_data_y, test_size=0.33, random_state=42)
```

```python
ecfp_final_data_train = pd.concat([ecfp_X_train,ecfp_y_train],axis=1)
ecfp_final_data_train.head()
```

```python
ecfp_final_data_test = pd.concat([ecfp_X_test,ecfp_y_test],axis=1)
```

```python
# import pycaret regression and init setup
from pycaret.regression import *
s = setup(ecfp_final_data_train, target = 'pIC50', session_id = 123)
```

```python
best = compare_models()
```

```python
best
```

```python
plot_model(best, plot = 'residuals')
```

```python
# plot error
plot_model(best, plot = 'error')
```

```python
# plot feature importance
plot_model(best, plot = 'feature')
```

```python
evaluate_model(best)
```

```python
# predict on test set
holdout_pred = predict_model(best,data=ecfp_X_test)
```
